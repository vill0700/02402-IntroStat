# Simulation from the Normal distribution

In this notebook we will work with Normal distribution, using the scipy.stats subpackage

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# For specific probability distributions (e.g., the Normal distribution) we will use a 
# new library: "Scipy" (actually only the subpackage "scipy.stats") 
import scipy.stats as stats
```

```python
np.set_printoptions(legacy='1.25')
```

```python
# "Draw" random variates (realizations/observations) from a Normal distribution with specified mean (mu) and standard deviation (sigma).

# first specify mu and sigma:
mu = 5
sigma = 2
print([mu, sigma])
```

We can get various information about the Normal distribution from stats.norm

To simulate "random variates" we use the .rvs method

```python
result = stats.norm.rvs(loc=mu, scale=sigma, size=1)
print(result)
```

Try repeating the cell above a few times and get new realizations each time. 

Do the results fit with your expactations (knowing mu and sigma)?

```python
# Lets try to "draw" 100 realizations in one go:
sample = stats.norm.rvs(loc=mu, scale=sigma, size=100)
print(sample)
```

```python
# make a histogram of the 100 observations:
plt.hist(sample)
plt.show()
```

Does this look like a Normal distribution?

Try repeating a few times. 

```python
# calculate the sample mean and sample standard deviation:
print([sample.mean(), sample.std(ddof=1)])
```

Are these "estimates" close to the expected values?

(they are "estimates" because they a calculated from the sample - i.e. from the data)
