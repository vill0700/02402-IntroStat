### E. 2.9 Continuous random variable

#### a\)

```python
import scipy.stats as stats
# for normal distribution standard parameters are loc=0, scale=1
# cdf(x, loc, scale)

print(stats.norm.cdf(2)) # P(Z<=2) for X~N(0,1)

print(stats.norm.cdf(2,1,1)) # P(X<=2) for X~N(1,1)

print(stats.norm.cdf(2,1,2)) # P(X<=2) for X~N(1,2)
```

```python
# Opret x-akse til alle plots
x = np.linspace(-6, 6, 1000)

# Opret figur og akser
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# 1: Standard normalfordeling (CDF)
ax1.plot(x, stats.norm.cdf(x, loc=0, scale=1))
ax1.set_title("Standard Normal CDF (=0, =1)")
ax1.grid()

# 2: Normalfordeling med mean=1, std=1 (CDF)
ax2.plot(x, stats.norm.cdf(x, loc=1, scale=1), color="orange")
ax2.set_title("Normal CDF (=1, =1)")
ax2.grid()

# 3: Normalfordeling med mean=1, std=2 (CDF)
ax3.plot(x, stats.norm.cdf(x, loc=1, scale=2), color="green")
ax3.set_title("Normal CDF (=1, =2)")
ax3.grid()

plt.tight_layout()
plt.show()
```

#### b\)

```python
print(stats.norm.ppf(stats.norm.cdf(2)))
```

#### c\)

```python
print(stats.norm.ppf(0.975))

print(stats.norm.ppf(0.975,1,1))

print(stats.norm.ppf(0.975,1,2))
```

```python
# Opret x-akse til alle plots
x = np.linspace(-6, 6, 1000)

# Opret figur og akser
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# 1: Standard normalfordeling (PpF)
ax1.plot(x, stats.norm.ppf(x, loc=0, scale=1))
ax1.set_title("Standard Normal PPF (=0, =1)")
ax1.grid()

# 2: Normalfordeling med mean=1, std=1 (PPF)
ax2.plot(x, stats.norm.ppf(x, loc=1, scale=1), color="orange")
ax2.set_title("Normal PPF (=1, =1)")
ax2.grid()

# 3: Normalfordeling med mean=1, std=2 (PPF)
ax3.plot(x, stats.norm.ppf(x, loc=1, scale=2), color="green")
ax3.set_title("Normal PPF (=1, =2)")
ax3.grid()

plt.tight_layout()
plt.show()
```
