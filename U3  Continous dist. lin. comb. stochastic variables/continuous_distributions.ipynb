```python
import scipy.stats as stats
import numpy as np

mu = 290
sigma = 4
```

```python
ans1 = stats.norm.sf(300, mu, sigma)  # sf(x, loc, scale)
print(ans1)
```

```python
ans2 = stats.norm.ppf(mu+1.96*sigma, mu, sigma), stats.norm.cdf(mu+1.96*sigma, mu, sigma)  # ppf(q, loc, scale), cdf(x, loc, scale)
print(ans2)
```
