### 3.2 Aluminium profile

```python
import numpy as np
import matplotlib as plt
import scipy.stats as stats
```

```python
""" Det er antaget stikprven kommer fra en population som normalfordelt s vi bruger t-dsistributionen"""
from numpy import mean


n = 16
mu = 180.05
sigma = 0.0959
ci = 0.90 # 90% konfindensinterval
alpha = 1 - ci
fraktil = 1 - (alpha/2)  #signifikansniveau
f = stats.t.ppf(fraktil,loc=0,df=n-1)
sem = sigma/np.sqrt(n) #standard error of the mean
print(f"{ci*100}% konfindensinterval for middelvrdien er: {mu-f*sem:.2f} <  < {mu+f*sem:.2f}")
```

```python
""" Det er antaget stikprven kommer fra en population som er normalfordelt s vi bruger Chi^2-distributionen"""
ci = 0.99 # 99% konfindensinterval
alpha = 1 - ci
sigma = 0.0959
fraktil = 1 - (alpha/2)  #signifikansniveau
# Quantiles of the chi-square distribution:
fraktil1, fraktil2 = stats.chi2.ppf([1-fraktil,fraktil],df=n-1)
print(f"{ci*100}% konfindensinterval for variansen er: {((n-1)*sigma**2)/fraktil2:.4f} <  < {((n-1)*sigma**2)/fraktil1:.4f}")
```

```python
muX = 3000
sigmaX = 3
n = 5
```

$X_i=N\sim(3000,3^2)$<br>
$Y=\sum\limits_{i=1}^{5} X_i$<br>

```python
muY = muX*n
sigmaY = n*(sigmaX**2)
print(f"Y~N({muY},{sigmaY})")
```

```python
maxdeviation = 20 #mm
prob = stats.norm.cdf(muY-maxdeviation, loc=muY, scale=np.sqrt(sigmaY))
print(f"Sandsynligheden for at afvige med meren end 2cm fra 15m er: {prob*2:.4f}")
```

```python
n = 9 # sample size
sample = np.array([3003,3005,2997,3006,2999,2998,3007,3005,3001])
mean = np.mean(sample) # sample mean 
std = np.std(sample,ddof=1) # sample standard deviation
sem = std/np.sqrt(n) # standard error of the mean
print(f"mean = {mean}, std = {std}, sem = {sem}")
```

SEM siger noget om afvigelsen af middelvrdien
STD siger noget om afvigelsen fra middelvrdien

```python
mu = mean
sigma = std
ci = 0.95 # 95% konfindensinterval
alpha = 1 - ci
fraktil = 1 - (alpha/2)  #signifikansniveau
f = stats.t.ppf(fraktil,loc=0,df=n-1)
sem = sigma/np.sqrt(n) #standard error of the mean
print(f"{ci*100}% konfindensinterval for middelvrdien er: {mu-f*sem:.2f} <  < {mu+f*sem:.2f}")
```

```python
""" Det er antaget stikprven kommer fra en population som normalfordelt s vi bruger t-dsistributionen"""

mu = mean
sigma = std
ci = 0.99 # 99% konfindensinterval
alpha = 1 - ci
fraktil = 1 - (alpha/2)  #signifikansniveau
f = stats.t.ppf(fraktil,loc=0,df=n-1)
sem = sigma/np.sqrt(n) #standard error of the mean
print(f"{ci*100}% konfindensinterval for middelvrdien er: {mu-f*sem:.2f} <  < {mu+f*sem:.2f}")
```

Det er strre, fordi vi vil sige noget med strre sikkerhed, men s bliver vi ogs ndt til at udvide, hvor meget fejl vi tillader.

```python
""" Det er antaget stikprven kommer fra en population som er normalfordelt s vi bruger Chi^2-distributionen"""
ci = 0.95 # 95% konfindensinterval
alpha = 1 - ci
fraktil = 1 - (alpha/2)  #signifikansniveau
# Quantiles of the chi-square distribution:
fraktil1, fraktil2 = stats.chi2.ppf([1-fraktil,fraktil],df=n-1)
print(f"{ci*100}% konfindensinterval for variansen er: {((n-1)*std**2)/fraktil2:.4f} <  < {((n-1)*std**2)/fraktil1:.4f}")
print(f"{ci*100}% konfindensinterval for standardafvigelsen er: {np.sqrt(((n-1)*std**2)/fraktil2):.4f} <  < {np.sqrt(((n-1)*std**2)/fraktil1):.4f}")
```

the mean is within the CI and the Variance is too, so while they cannot be certain the desired values are within the chosen CI

A larger value for n gives a better estimate of the mean and variance. Thus the true mean could be slightly larger and the true variation greater, but it is still within the same CI, so no new information is gained.
